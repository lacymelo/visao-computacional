```python
# Harris Corner Detection
# Para identificação de bordas na imagem
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# carregando a imagem
image = cv2.imread('../../assets/ovelha.png')
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# mostrando a imagem
plt.imshow(image)
```

```python
# carregando a imagem na cor cinza
grayImage = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
plt.imshow(grayImage, cmap='gray')
```

```python
# imagem real
realImage = cv2.imread('../../assets/ovelha-real.png')
realImage = cv2.cvtColor(realImage, cv2.COLOR_BGR2RGB)
plt.imshow(realImage)
```

```python
# imagem real em cinza
imageRealGray = cv2.cvtColor(realImage, cv2.COLOR_RGB2GRAY) 
plt.imshow(imageRealGray, cmap='gray')
```

```python
# converter RGB da imagem para float
convertFloatImageGray = np.float32(grayImage)
```

```python
# aplicando o algoritmo Harris Corner Detection
destinationImageGray = cv2.cornerHarris(src=grayImage, blockSize=2, ksize=3, k=0.04)
```

```python
# transformação morfológica pelo cv2.dilate()
destinationImageGray = cv2.dilate(destinationImageGray, None)

image[destinationImageGray > 0.01 * destinationImageGray.max()] = [255,255,0]
destinationImageGray.max()
```

```python
plt.imshow(image)
```

```python
# converter RGB da imagem para float
convertFloatImageRealGray = np.float32(imageRealGray)
```

```python
# aplicando o algoritmo Harris Corner Detection
destinationImageRealGray = cv2.cornerHarris(src=imageRealGray, blockSize=2, ksize=3, k=0.04)
```

```python
# transformação morfológica pelo cv2.dilate()
destinationImageRealGray = cv2.dilate(destinationImageRealGray, None)

realImage[destinationImageRealGray > 0.01*destinationImageRealGray.max()] = (255,255,0) 

destinationImageRealGray.max()
```

```python
plt.imshow(realImage)
```
