```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import cv2
```

```python
# fazendo leitura da imagem usando o OpenCV
image = cv2.imread('../../assets/da-6-meses.jpg')

plt.imshow(image)
```

```python
# corrigindo a coloração da imagem
correctingImage = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(correctingImage)
```

```python
# verificando o tamanho da imagem
correctingImage.shape
```

```python
# redimensionando a imagem
imageResizing = cv2.resize(correctingImage, (2064, 954))

plt.imshow(imageResizing)
```

```python
# ajustando a imagem em 50%
length = 0.5
width = 0.5

adjustedImage = cv2.resize(correctingImage, (0,0), correctingImage, width, length)
plt.imshow(adjustedImage)
```

```python
# rotacionando a imagem
rotatedImage = cv2.flip(correctingImage, -1)
plt.imshow(rotatedImage)
```

```python
# salvando a imagem girada
cv2.imwrite('../../assets/rotated-da.jpg', rotatedImage)
```

```python
# aumentando o tamanho da imagem
enlargingImage = plt.figure(figsize=(6,6))
ax = enlargingImage.add_subplot(111)
ax.imshow(adjustedImage)
```
