```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from PIL import Image
```

```python
# importando a imagem
image = Image.open('../../assets/da-filhote.png')

# image
```

```python
# transforma a imagem em um vetor
imageVector = np.asarray(image)
# type(imageVector)
imageVector.shape
```

```python
# apresenta a imagem vetorizada
plt.imshow(imageVector)
```

```python
# criando uma copia da imagem
imageCopy = imageVector.copy()

plt.imshow(imageCopy)
```

```python
plt.imshow(imageCopy[:,:,0], cmap='gray')
```
