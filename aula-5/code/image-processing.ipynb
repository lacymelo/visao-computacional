```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import cv2
```

```python
# fazendo a leitura da imagem com cv2
image = cv2.imread('../../assets/da-adulto.jpg')

# verificando tipo da imagem
type(image)

# verificando o tamanho do arquivo
image.shape
```

```python
plt.imshow(image)
```

Matplotlib trabalha com o RGB.
Open-cv trabalha com BGR.
Portanto é necessário corrigir a imagem

```python

# corrigindo a imagem
correctingImage = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(correctingImage)
```

```python
# carregando a imagem na escala cinza
imageScaleGray = cv2.imread('../../assets/da-adulto.jpg', cv2.IMREAD_GRAYSCALE)
plt.imshow(imageScaleGray, cmap='gray')
```
