```python
# Shi-Tomasi Corner Detection
# Para identificação de bordas na imagem
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# leitura da imagem
xadrez = cv2.imread('../../assets/xadrez.jpeg')
xadrez = cv2.cvtColor(xadrez, cv2.COLOR_BGR2RGB)
plt.imshow(xadrez)
```

```python
# imagem cinza
grayXadrez = cv2.cvtColor(xadrez, cv2.COLOR_BGR2GRAY)
plt.imshow(grayXadrez, cmap='gray')
```

```python
# aplicando o algoritmo
edges = cv2.goodFeaturesToTrack(grayXadrez, 100, 0.01, 10)
edges
```

```python
# converter para inteiro
edges = np.int0(edges)
edges
```

```python
# plotando os círculos na imagem
for i in edges:

    x,y = i.ravel()
    cv2.circle(xadrez, center=(x,y), radius=3, color=(0,255,0), thickness=10)

plt.imshow(xadrez)
```

```python
# leitura da imagem
realXadrez = cv2.imread('../../assets/xadrez-real.jpeg')
realXadrez = cv2.cvtColor(realXadrez, cv2.COLOR_BGR2RGB)
plt.imshow(realXadrez)
```

```python
# imagem cinza
xadrezGrayReal = cv2.cvtColor(realXadrez, cv2.COLOR_BGR2GRAY)
plt.imshow(xadrezGrayReal, cmap='gray')
```

```python
# aplicando o algoritmo
edgesReal = cv2.goodFeaturesToTrack(xadrezGrayReal, 500, 0.01, 10)
# convertendo para inteiro
edgesReal = np.int0(edgesReal)
```

```python
# plotando os círculos na imagem
for i in edgesReal:

    x,y = i.ravel()
    cv2.circle(realXadrez, center=(x,y), radius=3, color=(0,255,0), thickness=4)

plt.imshow(realXadrez)
```
