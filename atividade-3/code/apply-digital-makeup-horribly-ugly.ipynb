# Aplicar maquiagem digital (terrivelmente feia)

```python
from PIL import Image, ImageDraw
import face_recognition
```

```python
# exibir imagem
image = Image.open("../../assets/photograph/diego3g.jpeg")
# image
```

```python
# carregar o arquivo jpg em uma matriz numpy
image = face_recognition.load_image_file('../../assets/photograph/diego3g.jpeg')
```

```python
# encontrar todas as caracter√≠sticas faciais na imagem
face_landmarks_list = face_recognition.face_landmarks(image)
```

```python
# Cria uma imagem PIL a partir da matriz da imagem
pil_image = Image.fromarray(image)
# pil_image
```

```python
# percorre a lista de caracter√≠sticas faciais
for face_landmarks in face_landmarks_list:

    # inst√¢ncia um objeto PIL para desenhar na imagem
    d = ImageDraw.Draw(pil_image, 'RGBA')

    # Transformar as sobrancelhas em um pesadelo
    d.polygon(face_landmarks['left_eyebrow'], fill=(68, 54, 39, 128))
    d.polygon(face_landmarks['right_eyebrow'], fill=(68, 54, 39, 128))
    d.line(face_landmarks['left_eyebrow'], fill=(58, 54, 39, 150), width=5)
    d.line(face_landmarks['right_eyebrow'], fill=(58, 54, 39, 150), width=5)

    # Dar brilho aos l√°bios
    d.polygon(face_landmarks['top_lip'], fill=(150, 2, 0, 120))
    d.polygon(face_landmarks['bottom_lip'], fill=(150, 2, 0, 120))
    d.line(face_landmarks['top_lip'], fill=(150, 0, 0, 120), width=5)

    # Dar brilho aos olhos
    d.polygon(face_landmarks['left_eye'], fill=(0, 0, 255, 10))
    d.polygon(face_landmarks['right_eye'], fill=(0, 0, 255, 10))

    # Aplicar delineador
    d.polygon(face_landmarks['left_eye'] + [face_landmarks['left_eye'][0]], fill=(255, 0, 0, 110), width=6)
    d.polygon(face_landmarks['right_eye'] + [face_landmarks['right_eye'][0]], fill=(255, 0, 0, 110), width=6)


# Mostrar a imagem com as altera√ß√µes
pil_image
```
